Importing the libraries

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

Part 1 - Data Preprocessing

```python
train_datagen = ImageDataGenerator(rescale = 1./255, 
                                   shear_range = 0.2, 
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

Preprocessing test set

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

Building the CNN

Initializing the CNN

```python
cnn = tf.keras.models.Sequential()
```

Step 1 - Convolution filters

```python
cnn.add(tf.keras.layers.Conv2D(activation = 'relu', input_shape = (28, 28, 1), filters = 32, kernel_size = 3))
```

Step 2 - Pooling Size of the squre, stride the moving or sliding

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=4))
```

Adding a second convolutional layer, no need for input_shape for the next layers

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

Step 3 - Flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

Step 4 - Full Connection

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

Step 5 - Output Layer

```python
cnn.add(tf.keras.layers.Dense(units = 1, activation = 'softmax'))
```

Part 3 - Training the CNN

Compiling the CNN

```python
cnn.compile(optimizer = "adam", loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Training the CNN on Training set and evaluating it on test set

```python
cnn.fit(X_train, y_train, epochs = 25, validation_data = [X_test, y_test])
```
